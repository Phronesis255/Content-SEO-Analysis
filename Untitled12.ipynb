## Define weighting formulas

### Subtask:
Create two alternative weighting formulas to score the keywords based on 'Average Monthly Searches' and 'Competition Index'.


**Reasoning**:
Define two alternative weighting formulas as Python functions.



```python
# Formula 1: Simple ratio with inverse competition index and inverse URL count
# This formula gives a higher score for higher average monthly searches and lower competition index and lower URL count.
# It uses a simple ratio of searches to competition index and URL count, adding 1 to the competition index and URL count
# to avoid division by zero and smooth the result for very low values.
def weighting_formula_1(avg_monthly_searches, competition_index, url_count):
  return avg_monthly_searches / ((competition_index + 1) * (url_count + 1))

# Formula 2: Logarithmic scaling of searches and inverse of competition index and inverse URL count
# This formula uses a logarithmic scale for average monthly searches to reduce the impact
# of extremely high search volumes and multiplies it by the inverse of the competition index and URL count.
# A small constant (1) is added to the competition index and URL count to avoid division by zero.
import numpy as np
def weighting_formula_2(avg_monthly_searches, competition_index, url_count):
  return np.log(avg_monthly_searches + 1) * (1 / (competition_index + 1)) * (1 / (url_count + 1))
```

## Apply weighting formulas

### Subtask:
Apply the two weighting formulas to the filtered data to calculate a score for each keyword.


**Reasoning**:
Apply the defined weighting formulas to the filtered DataFrame to calculate the scores and display the result.



```python
df_filtered['Score Formula 1'] = df_filtered.apply(
    lambda row: weighting_formula_1(row['Average Monthly Searches'], row['Competition Index'], row['URL Count']), axis=1)

df_filtered['Score Formula 2'] = df_filtered.apply(
    lambda row: weighting_formula_2(row['Average Monthly Searches'], row['Competition Index'], row['URL Count']), axis=1)

display(df_filtered.head())
```
